```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import tifffile as tiff
import glob
from PIL import Image
import json
import rasterio as ras
```

```python
data_dir = r"mmflood/mmflood"
output_dir = os.path.join(data_dir, r"Output/Mask")  
os.makedirs(output_dir, exist_ok=True)

mask_files = []
for root, dirs, files in os.walk(data_dir):
    if "mask" in root.lower():  
        for file in files:
            if file.endswith(".tif") or file.endswith(".tiff"):
                mask_files.append(os.path.join(root, file))

print(f"Found {len(mask_files)} Mask files.")
```

```python
for mask_path in mask_files:
    print(f"\nProcessing: {mask_path}")

    with ras.open(mask_path) as mask:
        mask_data = mask.read(1)  
        profile = mask.profile  

    relative_path = os.path.relpath(mask_path, data_dir)
    relative_folder = os.path.dirname(relative_path)  
    save_folder = os.path.join(output_dir, relative_folder)  

    os.makedirs(save_folder, exist_ok=True)
    save_path = os.path.join(save_folder, os.path.basename(mask_path).replace(".tif", ".png"))
    
    plt.figure(figsize=(10, 6))
    plt.imshow(mask_data, cmap="gray")  
    plt.colorbar(label="Mask Values")
    plt.title(f"Mas Data: {os.path.basename(mask_path)}")
    plt.savefig(save_path, dpi=300, bbox_inches='tight') 
    plt.close()  
    print(f"Saved: {save_path}")
```
