```python
data_dir = r"mmflood/mmflood"
output_dir = os.path.join(data_dir, r"Output/Hydro")  
os.makedirs(output_dir, exist_ok=True)

hydro_files = []
for root, dirs, files in os.walk(data_dir):
    if "hydro" in root:  
        for file in files:
            if file.endswith(".tif") or file.endswith(".tiff"):
                hydro_files.append(os.path.join(root, file))

print(f"Found {len(hydro_files)} hydro files.")
```

```python
for hydro_path in hydro_files:
    print(f"\nProcessing: {hydro_path}")

    with ras.open(hydro_path) as hydro:
        hydro_data = hydro.read(1)  
        profile = hydro.profile  

    relative_path = os.path.relpath(hydro_path, data_dir)
    relative_folder = os.path.dirname(relative_path)  
    save_folder = os.path.join(output_dir, relative_folder)  

    os.makedirs(save_folder, exist_ok=True)
    save_path = os.path.join(save_folder, os.path.basename(hydro_path).replace(".tif", ".png"))
    
    plt.figure(figsize=(10, 6))
    plt.imshow(hydro_data, cmap="Blues")  
    plt.colorbar(label="Hydro Data Intensity")
    plt.title(f"Hydrological Data: {os.path.basename(hydro_path)}")
    plt.savefig(save_path, dpi=300, bbox_inches='tight')  
    plt.close()  
    print(f"Saved: {save_path}")
```
