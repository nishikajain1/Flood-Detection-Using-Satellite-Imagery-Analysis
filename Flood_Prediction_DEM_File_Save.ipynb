```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import tifffile as tiff
import glob
from PIL import Image
import json
```

```python
import rasterio as ras
```

```python
data_dir = r"mmflood/mmflood"
output_dir = os.path.join(data_dir, r"Output/DEM")  
os.makedirs(output_dir, exist_ok=True)

dem_files = []
for root, dirs, files in os.walk(data_dir):
    if "dem" in root.lower():  
        for file in files:
            if file.endswith(".tif") or file.endswith(".tiff"):
                dem_files.append(os.path.join(root, file))

print(f"Found {len(dem_files)} DEM files.")
```

```python
for dem_path in dem_files:
    print(f"\nProcessing: {dem_path}")

    with ras.open(dem_path) as dem:
        dem_data = dem.read(1)  
        profile = dem.profile  

    relative_path = os.path.relpath(dem_path, data_dir)
    relative_folder = os.path.dirname(relative_path)  
    save_folder = os.path.join(output_dir, relative_folder)  

    os.makedirs(save_folder, exist_ok=True)
    save_path = os.path.join(save_folder, os.path.basename(dem_path).replace(".tif", ".png"))
    
    plt.figure(figsize=(10, 6))
    plt.imshow(dem_data, cmap="terrain")  
    plt.colorbar(label="Elevation (m)")
    plt.title(f"DEM Data: {os.path.basename(dem_path)}")
    plt.savefig(save_path, dpi=300, bbox_inches='tight') 
    plt.close()  
    print(f"Saved: {save_path}")
```
