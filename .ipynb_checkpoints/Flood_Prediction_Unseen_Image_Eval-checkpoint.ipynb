```python
import os
import numpy as np
import cv2
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.models import load_model
model = load_model(r"mera_Unet_flood_model.keras")
```

```python
test_image_path = r"C:\Users\dhruv\Downloads\OIP (3).jpeg"
test_image = cv2.imread(test_image_path)
test_image = cv2.resize(test_image, (512, 512))  
test_image = test_image / 255.0  
test_image = np.expand_dims(test_image, axis=0)  
predicted_mask = model.predict(test_image)
predicted_mask = predicted_mask.squeeze()
flood_area = np.mean(predicted_mask)
flood_area_coverage = ((flood_area * 100)-1)*100
if flood_area_coverage < 0:
    flood_area_coverage += 100
elif (flood_area_coverage > 100):
    flood_area_coverage -= 100
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(cv2.imread(test_image_path), cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(predicted_mask, cmap="gray")
plt.title("Predicted Mask")
plt.axis("off")
plt.show()
print(f"Flooded Area Coverage: {flood_area_coverage:.2f}%")
```

```python
flood_area_actual = flood_area*100
print(f"Flooded Area: {flood_area_actual:.2f}%")
```
